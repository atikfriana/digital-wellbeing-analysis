Cleaning survey data

```python
survey_data=df
survey_data.head()
```

```python
survey_data.columns = survey_data.columns.str.strip()  # Removes leading/trailing spaces
survey_data.rename(columns={'Average screen time on daily basis (phone or laptop)':'Screen time',
                      'Average time spent on social media on daily basis':'Social Media usage',
                       'What\'s your most used social media platform?':'Fav social media',
                       'Your content-format preference while using Instagram':'Format preference Instagram',
                       'Your content-format preference while using Youtube':'Format preference Youtube',
                       'What kind of content do you prefer?':'Content preference',
                       'Do you feel that consuming short-form content has reduced your ability to focus on longer tasks or content?':'Reduced focus(Yes/No)?',
                       'How long can you stay focused on a task without getting distracted?':'Duration of focus',
                       'Do you find it difficult to focus on longer videos or read articles after consuming short-form content?':'Difficulty focusing(Yes/No)?',
                       'Do you feel addicted to these platforms and short-length contents?':'Addiction (Yes/No)?',
                       'How do you feel after spending time watching short-form videos?':'Happy/Guilty/Relaxed?',
                       'Have you ever tried to reduce the time spent on short-form videos?':'Attempt at reducing addiction (Yes/No)',
                       'Has it affected your productivity in terms of work and life?':'Effect on productivity(Yes/No)?'
                      },inplace=True)
survey_data.head()
```

```python
survey_data['Timestamp']=pd.to_datetime(survey_data['Timestamp'])
survey_data['Gender']=survey_data['Gender'].astype('category')
survey_data['Age group']=survey_data['Age group'].astype('category')
survey_data['Profession']=survey_data['Profession'].astype('category')
survey_data['Screen time']=survey_data['Screen time'].astype('category')
survey_data['Fav social media']=survey_data['Fav social media'].astype('category')
survey_data['Format preference Instagram']=survey_data['Format preference Instagram'].astype('category')
survey_data['Format preference Youtube']=survey_data['Format preference Youtube'].astype('category')
survey_data['Content preference']=survey_data['Content preference'].astype('category')
survey_data['Reduced focus(Yes/No)?']=survey_data['Reduced focus(Yes/No)?'].astype('category')
survey_data['Duration of focus']=survey_data['Duration of focus'].astype('category')
survey_data['Difficulty focusing(Yes/No)?']=survey_data['Difficulty focusing(Yes/No)?'].astype('category')
survey_data['Addiction (Yes/No)?']=survey_data['Addiction (Yes/No)?'].astype('category')
survey_data['Happy/Guilty/Relaxed?']=survey_data['Happy/Guilty/Relaxed?'].astype('category')
survey_data['Attempt at reducing addiction (Yes/No)']=survey_data['Attempt at reducing addiction (Yes/No)'].astype('category')
survey_data['Effect on productivity(Yes/No)?']=survey_data['Effect on productivity(Yes/No)?'].astype('category')
survey_data.info()
```

```python
survey_data.isna().sum()
```

```python
survey_data['Content preference']=survey_data['Content preference'].fillna(survey_data['Content preference'].mode()[0])
survey_data.isna().sum()
```
