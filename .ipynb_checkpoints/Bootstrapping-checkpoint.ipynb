```python
import pandas as pd

# Step 1: Load the original data (replace with the path to your file)
data = pd.read_excel('Dataset.xlsx')  # Replace 'your_survey_data.xlsx' with your actual file path

# Step 2: Perform bootstrapping (sample 1000 records with replacement)
survey = data.sample(n=1000, replace=True, random_state=42)

# Step 3: Check the first few rows of the bootstrapped data to confirm
print(survey.head())
survey.to_csv('survey.csv',index=False)

# Step 4: Save the new bootstrapped dataset to an Excel or CSV file
```
